# Odd zeta PSLQ demo

Quick run scanning $\zeta(3)$ for algebraic relations up to degree 10.

```python
from z_p.experiments.maximal_pslq_scan import run_maximal_scan
import mpmath as mp

run_maximal_scan(
    "Zeta(3)",
    lambda: mp.zeta(3),
    max_degree=10,
    max_coeff=5000,
    precision_dps=400,
)
```

Tune `max_degree`, `max_coeff`, and `precision_dps` to widen the search. Expected outcome for conjecturally transcendental constants is `None` (no relation) within the searched bounds.
